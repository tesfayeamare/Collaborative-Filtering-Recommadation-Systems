```python
import numpy as np 
import pandas as pd 
from datetime import datetime
import ast
import random

import os
```

```python
df_2000s = pd.read_csv("/kaggle/input/spotify-top-2000s-mega-dataset/Spotify-2000.csv")
df_2000s.head()
```

```python
df_2000s.rename(columns={'Song Name':'Title'},inplace=True)
df_2000s = df_2000s.drop('Index', axis=1)
df_2000s.head()
```

```python
temp_genre_list = []
temp_duration_list = []
temp_energy_list = []
temp_danceability_list = []
temp_liveness_list = []
temp_valence_list = []
temp_acousticness_list = []
temp_speechiness_list = []
temp_loudness_list = []
for index, row in df_2000s.iterrows():
    inner_list = [row['Top Genre']]
    temp_genre_list.append(inner_list)
    temp_duration_list.insert(len(temp_duration_list), int(row['Length (Duration)'].replace(',', '')) * 1000)
    temp_energy_list.insert(len(temp_energy_list), int(row['Energy']) / 100)
    temp_danceability_list.insert(len(temp_danceability_list), int(row['Danceability']) / 100)
    temp_liveness_list.insert(len(temp_liveness_list), int(row['Liveness']) / 100)
    temp_valence_list.insert(len(temp_valence_list), int(row['Valence']) / 100)
    temp_acousticness_list.insert(len(temp_acousticness_list), int(row['Acousticness']) / 100)
    temp_speechiness_list.insert(len(temp_speechiness_list), int(row['Speechiness']) / 100)
    temp_loudness_list.insert(len(temp_loudness_list), format(row['Loudness (dB)'], '.3f'))
df_2000s['Genre'] = temp_genre_list

df_2000s['Duration (ms)'] = temp_duration_list
df_2000s['Energy'] = temp_energy_list
df_2000s['Danceability'] = temp_danceability_list
df_2000s['Liveness'] = temp_liveness_list
df_2000s['Valence'] = temp_valence_list
df_2000s['Acousticness'] = temp_acousticness_list
df_2000s['Speechiness'] = temp_speechiness_list
df_2000s['Loudness (dB)'] = temp_loudness_list
df_2000s = df_2000s.drop(['Top Genre', 'Length (Duration)', 'Beats Per Minute (BPM)'], axis=1)
df_2000s.head()
```

```python
df_2021 = pd.read_csv("/kaggle/input/spotify-top-200-charts-20202021/spotify_dataset.csv")
# df_2021.head()
```

```python
from faker import Faker
```

```python
users = []
total_users = 100

fake = Faker('en_US')
sex_list = ['F', 'M', 'NA']
val = 1

for i in range(total_users):
    users.append({
            "user_id" : 'U' + str(val).zfill(3),
            "user_name" : fake.name(),
            "city" : fake.city(),
            "country" : fake.country(),
            "age" : random.randint(10,70),
            "occupation" : fake.job(),
            "sex" : np.random.choice(sex_list)
            })
    val = val + 1
        
    
df_user = pd.DataFrame(users)
# df_user.insert(0, 'user_id', range(1, 1 + len(df_user)))
df_user.head()
```

# Create Query Dataset

```python
song_genre = ['pop', 'rock', 'hip hop', 'r&b', 'dance pop', 'edm', 'pop dance', 'tropical house', 'uk dance', 'modern rock']
queries = []
val = 1

for i,user in df_user.iterrows():
    queries.append({
        "query_id" : 'Q' + str(val).zfill(4),
        "cond_01" : "country=" + user['country']
        })
    val = val + 1
    
    queries.append({
        "query_id" : 'Q' + str(val).zfill(4),
        "cond_01" : "age=" + str(user['age'])
        })
    val = val + 1
    
    queries.append({
        "query_id" : 'Q' + str(val).zfill(4),
        "cond_01" : "genre=" + np.random.choice(song_genre)
        })
    val = val + 1
    
    queries.append({
        "query_id" : 'Q' + str(val).zfill(4),
        "cond_01" : "country=" + user['country'],
        "cond_02" : "genre=" + np.random.choice(song_genre)
        })
    val = val + 1
    
    queries.append({
        "query_id" : 'Q' + str(val).zfill(4),
        "cond_01" : "age=" + str(user['age']),
        "cond_02" : "genre=" + np.random.choice(song_genre)
        })
    val = val + 1
    
    queries.append({
        "query_id" : 'Q' + str(val).zfill(4),
        "cond_01" : "country=" + user['country'],
        "cond_02" : "age=" + str(user['age'])
        })
    val = val + 1
    
    queries.append({
        "query_id" : 'Q' + str(val).zfill(4),
        "cond_01" : "country=" + user['country'],
        "cond_02" : "age=" + str(user['age']),
        "cond_03" : "genre=" + np.random.choice(song_genre)
        })
    val = val + 1
    
    
df_query = pd.DataFrame(queries)
print(df_query.shape)
# df_query.head(20)
```

```python
df_query.drop_duplicates(subset=["cond_01", "cond_02", "cond_03"], keep=False, inplace=True)
print(df_query.shape)
df_query.head(20)
```

```python
df_query.to_csv('query.csv')
```

```python
query_ratings = []
total_records = 50000

for i in range(total_records):

    query_ratings.append({
            "user_id" : np.random.choice(df_user['user_id']),
            "query_id": np.random.choice(df_query['query_id']),
            "rating": random.randint(1,100)
            })


df_query_ratings = pd.DataFrame(query_ratings)
print(df_query_ratings.shape)
```

```python
df_query_ratings.drop_duplicates(subset=["user_id", "query_id"], keep=False, inplace=True)
print(df_query_ratings.shape)
df_query_ratings.head(10)
```

```python
df_query_ratings.to_csv('query_ratings.csv')
```
